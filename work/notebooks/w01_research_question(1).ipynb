## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Reading the data directly from the raw GitHub repository link
url = "https://raw.githubusercontent.com/mutallimov7/Flyrank-ML-Internship/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Extracting 3 real numbers to support the lane choice
total_pages = len(df)
declining_pages = len(df[df["trend_direction"] == "down"])
avg_impressions = df["impressions_90d"].mean()

print(f"1. Total number of pages in the dataset: {total_pages}")
print(f"2. Number of pages with declining traffic (potential targets): {declining_pages}")
print(f"3. Average 90-day impressions per page: {avg_impressions:.0f}")
```
