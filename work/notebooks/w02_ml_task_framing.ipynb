## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My chosen lane is Lane 2: Refresh / Content Opportunity Scoring. In ML terms, this is a Scoring and Ranking task built on top of a Binary Classification model. The system classifies whether a page is declining or missing opportunities, and uses that predicted probability to rank the pages by priority.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

Since we don't have perfect future labels in the starter data, the proxy target is identifying current decline: is_declining. We derive this label directly from the trend_direction == 'down' observable signal.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

The primary success metric is Precision@K (e.g., Precision@50). The content review team has limited capacity. We care deeply about ensuring that if the model recommends 50 pages to fix, the vast majority of them are genuinely high-impact opportunities, rather than trying to achieve general accuracy across millions of irrelevant pages.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

# Reading the data directly from your GitHub repository link
url = "https://raw.githubusercontent.com/mutallimov7/Flyrank-ML-Internship/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

# Creating the proxy target column
df['target_is_declining'] = (df['trend_direction'] == 'down').astype(int)

print(f"Unit of analysis: One row = one content item (page).")
print(f"Total rows in dataset: {len(df)}\n")
print("Here is what the target column looks like alongside the data:")
print(df[['content_id', 'impressions_90d', 'trend_direction', 'target_is_declining']].head())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed SQL rule (e.g., "refresh pages older than 180 days with >500 impressions") is rigid. It creates hard cutoffs where a 179-day-old page is ignored while a 181-day-old page is flagged. ML evaluates the continuous interaction of multiple features (age, position, CTR, engagement) simultaneously to find non-obvious decay patterns that a simple rule would miss.
